```python
import pandas as pd
import cufflinks as cf
import plotly.offline as plyo
```

```python
plyo.init_notebook_mode(connected=True)
```

```python
import numpy as np
```

```python
# Creating a 5D array of 250 rows and 5 columns using random.standard_normal data
```

```python
a = np.random.standard_normal((250,5)).cumsum(axis=0)
```

```python
index = pd.date_range('2019-1-1', freq='B', periods=len(a)) #creating index for dataframe, freq b is business daily
```

```python
df = pd.DataFrame( 100 + 5 * a, columns=list('abcde'), index=index) #creating dataframe and storing array with multiplications
```

```python
df.head()
```

```python
plyo.iplot(df.iplot(asFigure=True), 
           image = 'png', 
           filename='ply_01')
```

```python
# Different parameters to customize the plot
```

```python
plyo.iplot(
            df[['a','b']].iplot(asFigure=True,
                               theme='polar',
                               title='A Time Series Plot',
                               xTitle='date',
                               yTitle='value',
                               mode={'a': 'markers','b': 'lines+markers'},
                               symbol={'a': 'circle','b': 'diamond'},
                               size=3.5,
                               colors={'a': 'blue', 'b': 'magenta'}),
            image='png',
            filename='ply_02')
```

```python
# Creation of histogram through plyo.iplot
```

```python
plyo.iplot(
            df.iplot(kind='hist',
                    subplots=True,
                    bins=15,
                    asFigure=True),
            image='png',
            filename='ply_03')
```
